## Introduction to tensors

```python
# Import tensorflow
import tensorflow as tf
print(tf.__version__)
```

```python
# Create tensors with tf.constant()
scalar = tf.constant(7)
scalar
```

```python
# Check the number of dimensions of a tensor (ndim stands for number of dimensions)
scalar.ndim
```

```python
# Create a vector
vector = tf.constant([10, 10])
vector
```

```python
# Check the dimensions of vector
vector.ndim
```

```python
# Create a matrix that has more than 1 dimension
matrix = tf.constant([[10, 7],
                      [7, 10]])
matrix
```

```python
# Check the dimensions of matrix
matrix.ndim
```

```python
# Create another matrix
another_matrix = tf.constant([[10., 9.],
                              [9., 10.],
                              [8., 9.]], 
                             dtype=tf.float16  # Specify the datatype
                             )
another_matrix
```

```python
# Create a tensor
tensor = tf.constant([[[1, 2, 3],
                       [4, 5, 6]],
                      [[1, 2, 3],
                       [4, 5, 6]]])
tensor
```

```python
# Check the number of dimensions of the tensor
tensor.ndim
```

* Scalar: a single number
* vector: a number with direction
* matrix: a 2 dimensional array of numbers
* tensor: an n-dimensional array of numbers (n can be 0, 1, 2, 3, 4, 5)

### Creating Random Tensors

```python
# Create a random (but the same) tensor
random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal(shape=(3, 2))
random_1
```

```python
# Create another random tensor
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3, 2))
random_2
```

```python
# Check to see if random_1 and random_2 are equal
random_1 == random_2
```

### Shuffle the order of elements in a tensor

```python
# Shuffle a tensor
not_shuffled = tf.constant([[10, 7],
                            [3, 4],
                            [2, 5]])
not_shuffled.ndim
```

```python
not_shuffled
```

```python
# Shuffle non_shuffled with operation level seed set to 42
tf.random.shuffle(not_shuffled, seed=42)  # Operation level seed
```

```python
# Shuffle non_shuffled again with operation level seed set to 42
tf.random.shuffle(not_shuffled, seed=42)
```

```python
# Shuffle non_shuffled with global level seed set to 42
tf.random.set_seed(42)
tf.random.shuffle(not_shuffled)
```

```python
# Shuffle non_shuffled again with global level seed set to 42
tf.random.set_seed(42)
tf.random.shuffle(not_shuffled)
```

```python
tf.random.set_seed(42)
tf.random.shuffle(not_shuffled, seed=42)
```

```python
tf.random.set_seed(21)
tf.random.shuffle(not_shuffled, seed=42)
```

```python
tf.random.set_seed(21)
tf.random.shuffle(not_shuffled, seed=42)
```

### Manipulating tensors (tensor operations)

**General Operations**

```python
# Add values to a tensor
tensor = tf.constant([[10, 7],
                      [3, 4]])
tensor + 10
```

```python
tf.add(tensor, 10)
```

```python
# Multiply tensor by a number
tensor * 10
```

```python
tf.multiply(tensor, 10)
```

```python
# Sustract a number from the tensor
tensor - 10
```

**Matrix Multiplication**

```python
tf.linalg.matmul(tensor, tensor)
```

```python
tensor * tensor
```

```python
tensor @ tensor
```

```python
A = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])
A
```

```python
B = tf.constant([[1, 2, 3],
                 [4, 5, 6],
                 [7, 8, 9]])
B
```

```python
# tf.linalg.matmul(A, B)  --> ERROR!
tf.matmul(tf.transpose(A), B)
```

**The dot product**

```python
tf.tensordot(tf.transpose(A), B, axes=1)
```

### Aggregating tensors

```python
A = tf.constant(np.random.randint(0, 100, size=50))
```

```python
A
```

```python
# Find the maximum
tf.reduce_min(A)
```

```python
# Find the maximum
tf.reduce_max(A)
```

```python
# Find the mean
tf.reduce_mean(A)
```

```python
# Find the sum
tf.reduce_sum(A)
```

```python
# Find the variance
tf.math.reduce_variance(tf.cast(A, dtype=tf.float32))
```

```python
# Find the std
tf.math.reduce_std(tf.cast(A, dtype=tf.float32))
```

### Find the positional max or min

```python
# Create an new tensor for finding the positional minimum and maximum
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# Find the positional maximum
tf.argmax(F)
```

```python
F[tf.argmax(F)]
```

```python
tf.reduce_max(F) == F[tf.argmax(F)]
```

```python
# Find the positional minimum
tf.argmin(F)
```

```python
# Find the minimum using the positional argmin index
F[tf.argmin(F)]
```

### One-Hot encoding tensors

```python
# Create a list of indices
some_list = [0, 1, 2, 3]  # Could be red, green, blue, purple

# One hot encode our list of indices
tf.one_hot(some_list, depth=4)
```

```python
# Specify custom values for one-hot-encoding
tf.one_hot(some_list, depth=4, on_value='ON', off_value='OFF')
```
